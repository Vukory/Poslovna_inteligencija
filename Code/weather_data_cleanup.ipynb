```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the data
df = pd.read_csv('weather_data.csv')

# Step 2: Convert the 'utc_timestamp' column to datetime
df['utc_timestamp'] = pd.to_datetime(df['utc_timestamp'])

# Step 3: Set the 'utc_timestamp' as the index
df.set_index('utc_timestamp', inplace=True)

# Step 4: Create a boolean DataFrame indicating missing values
missing_values = df.isnull()

# Step 5: Plot the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(missing_values, cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values in Weather Data')
plt.xlabel('Columns')
plt.ylabel('Timestamps')
plt.show()
```

```python
# Step 6: Resample the data to monthly frequency and calculate the mean for each month
df_monthly = df.resample('M').mean()

# Step 7: Save the cleaned and resampled data to a new CSV file (optional)
df_monthly.to_csv('weather_data_monthly.csv')

# Display the first few rows of the cleaned data
df_monthly.head()
```

```python
# Step 8: Extract the country codes from the column names
columns = df.columns
country_codes = set([col.split('_')[0] for col in columns if '_' in col])

# Step 9: Count the number of unique countries
num_countries = len(country_codes)

print(f"There are {num_countries} unique countries in the dataset.")

# Print the list of country codes
print("List of country codes:")
print(sorted(country_codes))
```
